# Business Problem - Predict the price of bangalore house

### using SUPPORT VECTOR REGRESION - Supervised ML Algorithm

```python
import pandas as pd
```

```python
#Loading Datasets >> Data is already cleaned/preprocessed

df = pd.read_csv(r"https://drive.google.com/uc?export=download&id=1xxDtrZKfuWQfl-6KA9XEd_eatitNPnkB")
df.head()                 
```

```python
X = df.drop('price',axis=1)
y = df['price']

print('shape of X = ', X.shape)
print('shape of y = ', y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.1 , random_state = 23)

print('shape of X_train = ',  X_train.shape)
print('shape of X_test = ',  X_test.shape)
print('shape of y_train = ',  y_train.shape)
print('shape of y_test = ',  y_test.shape)
```

```python
#Feature Scaling

from sklearn.preprocessing import StandardScaler

sc= StandardScaler()
sc.fit(X_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

#### Support Vector Regression - ML Model Training

```python
#ML Model Training

from sklearn.svm import SVR
```

```python
svr_rbf = SVR(kernel = 'rbf')
svr_rbf.fit(X_train , y_train)
svr_rbf.score(X_test , y_test)
```

```python
svr_lin = SVR(kernel = 'linear')
svr_lin.fit(X_train , y_train)
svr_lin.score(X_test , y_test)
```

```python
svr_poly= SVR(kernel = 'poly' , degree=2)
svr_poly.fit(X_train , y_train)
svr_poly.score(X_test , y_test)
```

Now looking at the above results we see that the linear regression suits best to our model and the rest are non efficient.

```python
X_test[0]
```

```python
#Home Price Prediction

svr_lin.predict([X_test[0]])
```

```python
y_pred = svr_lin.predict(X_test)
y_pred
```

```python
y_test
```

```python
#rms and rmse

from sklearn.metrics import mean_squared_error
import numpy as np

mse= mean_squared_error(y_test , y_pred)
rmse = np.sqrt(mse)

print('MSE = ', mse)
print('RMSE = ', rmse)
```
