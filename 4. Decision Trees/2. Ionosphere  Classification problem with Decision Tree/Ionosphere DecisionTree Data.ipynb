```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('ionosphere.data', header=None)
```

```python
df
```

```python
df.groupby(34).count()
```

```python
class_b = df[df[34] == 'b']
```

```python
class_b_over = class_b.sample(df[df[34] == 'g'].count()[0], replace=True)
```

```python
class_g = df[df[34] == 'g']
```

```python
#Now we are creating a new df with same name so that we don't need to change the code further

df = pd.concat([class_b_over, class_g])
df
```

```python
X = df.iloc[:,:-1]
Y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test , y_train, y_test  = train_test_split(X,Y, train_size=0.8, random_state=23)
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(criterion='gini')
```

```python
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
model.predict(X_test)
```
